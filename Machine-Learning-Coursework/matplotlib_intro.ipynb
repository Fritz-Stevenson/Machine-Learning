## Matplotlib Plotting Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#obtain data
import numpy as np
x = [i for i in range(10)]
y = np.random.normal(10,2,10)

#constuct figure and insert data into subplot
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y)

#customize figure
ax.set(title='Sequential Normal Distribution', xlabel='Sequence Index', ylabel='Value')
fig.savefig('matplotlib_example.png')
;
```

```python
#Bar Charts
fig, ax = plt.subplots()
x = [i for i in range(1,10)]
y = [list(np.random.randint(1,10,1000)).count(i) for i in range(10)]
ax.bar(x,y)
ax.set(title='Occurence of Numbers Across Random Dataset')
plt.xticks(x);
```

```python
#Histogram
fig, ax = plt.subplots()
x = [np.sqrt(i) for i in np.random.randn(500)]
ax.hist(x)
plt.show;
```

## Pyplot vs Matplotlib's oo(object-oriented) Approach

* pyplot is suited for quick exploratory plots
* OO plotting allows more customization

```python
# Setting a slice of the dataframe- Only men
heart_disease.head()
# After some googling, https://archive.ics.uci.edu/ml/datasets/Heart+Disease/, value 1 is male
male_df = heart_disease[heart_disease['sex']==1]
# Chest pain histogram
male_df.cp.plot.hist(bins=5);
```

```python
# Assisted OO method

fig, ax = plt.subplots(figsize=(6,10))
male_df.plot.scatter(figsize=(15,15),x='age', y='chol', c='target', ax=ax)
ax.set(xlabel='Age', ylabel='Cholesterol', title='Male Cholesterol by Age')
plt.show();
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(12,6))
# Setting customizable label colors 
from matplotlib import colors
cmap, norm = colors.from_levels_and_colors([0, 1], ['teal', 'red'], extend='max')
scatter = ax.scatter(x=male_df['age'], 
                     y=male_df['chol'],
                    c=male_df['target'],
                    cmap=cmap, norm=norm)
ax.set(xlabel='Age', ylabel='Cholesterol', title='Male Cholesterol by Age')
ax.legend(*scatter.legend_elements(), title='target')
ax.plot(male_df['age'],[male_df['chol'].mean() for i in range(len(male_df))] , linestyle='dotted')
leg = ax.get_legend()
;
```

```python
#cigarrettes per day by sex for those with heart disease

hd_df = heart_disease[heart_disease['target']==1]
fig, ax = plt.subplots(figsize=(12,10))
# Setting customizable label colors 
from matplotlib import colors
cmap, norm = colors.from_levels_and_colors([0, 1], ['pink', 'blue'], extend='max')
scatter = ax.scatter(x=hd_df['age'], 
                     y=hd_df['trestbps'],
                    c=hd_df['sex'],
                    cmap=cmap, norm=norm)
ax.set(xlabel='Age', ylabel='Resting Heart Rate', title='Resting Heart Rate For Patients with Heart Disease')
ax.legend(*scatter.legend_elements(), title='sex')
ax.axhline(y=heart_disease['trestbps'].mean(),
           linestyle='dotted', 
           label='mean cigs over all individuals',
          color='red')
leg = ax.get_legend()
;
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(10,6), sharex=True)
over_45 = heart_disease[heart_disease['age']>45]
# 2 scatter plots for those over 45yo: thalach and cholesterol 
cmap1, norm1 = colors.from_levels_and_colors([0, 1], ['purple', 'yellow'], extend='max')
cmap2, norm2 = colors.from_levels_and_colors([0, 1], ['blue', 'pink'], extend='max')

scatter1= ax1.scatter(x=over_45['age'], y=over_45['chol'], c=over_45['target'], cmap=cmap1, norm=norm1)
ax1.set(title='Cholesterol by Age', xlabel='Age', ylabel='Cholesterol Level')
ax1.legend(*scatter1.legend_elements(), title='Heart Disease')
ax1.axhline(over_45['chol'].mean(), linestyle='-')
scatter2= ax2.scatter(x=over_45['age'], y=over_45['thalach'],c=over_45['target'], cmap=cmap2, norm=norm2)
ax2.set(title='Maximum Heart Rate by Age', xlabel='Age', ylabel='Maximum Heart Rate')
ax2.axhline(over_45['thalach'].mean(), linestyle='dashdot')
ax2.legend(*scatter2.legend_elements(), title='Heart Disease');
fig.suptitle('Heart Disease Analysis (Over 45)');
```
